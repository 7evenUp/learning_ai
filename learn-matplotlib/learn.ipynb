```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

### Basic usage with subplots

```python
# simple plot
xAxis = np.array([2, 4, 6, 8, 9])
yAxis = np.array([2, 5, 4, 8, 6])

fig, ax = plt.subplots(figsize=(5, 3)) # numbers in figsize: 8 refers to width, 4 refers to height of plot image itself
ax.plot(xAxis, yAxis)
ax.set_xlabel('Ось X')
ax.set_ylabel('Ось Y')
ax.set_title('Title')
```

```python
# Scatter plot
xAxis = np.arange(10)
yAxis = np.arange(10)

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(xAxis, yAxis, s=50, facecolor='yellow', edgecolor='black')
```

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
# the histogram of the data
n, bins, patches = ax.hist(x, 75, density=True, facecolor='green', alpha=0.25)

ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Aardvark lengths\n (not really)')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')
ax.annotate('local max', xy=(115, 0.025), xytext=(140, 0.025),
            arrowprops=dict(facecolor='black', shrink=0.05, alpha=0.75))
ax.axis([55, 175, 0, 0.03])
ax.grid(True)
```
